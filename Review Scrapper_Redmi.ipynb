```python
from bs4 import BeautifulSoup
import requests

URL="https://www.flipkart.com/redmi-note-8-pro-shadow-black-128-gb/p/itm2f31dd5ad9153?pid=MOBFHGTDDTQ4CZVZ&lid=LSTMOBFHGTDDTQ4CZVZQHFGIA&marketplace=FLIPKART&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_1_10_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_10_na_na_na&fm=SEARCH&iid=8fe26b20-d651-4921-b78e-5782aa889551.MOBFHGTDDTQ4CZVZ.SEARCH&ppt=sp&ppn=sp&ssid=ruto2751s00000001587308520174&qH=6355865b5e13ddee"
```

```python
req_data = requests.get(URL)
review_soup = BeautifulSoup(req_data.content, 'html.parser')
```

```python
baseUrl = "https://www.flipkart.com"
all_reviews_link = review_soup.find('div', {'class': 'swINJg _3nrCtb'})

all_reviews_link.find_parent()
data = str(all_reviews_link.find_parent().get('href'))
print(baseUrl)
print(data)

new_url = baseUrl + data
```

```python
req_data = requests.get(new_url)
review_soup = BeautifulSoup(req_data.content, 'html.parser')

all_reviews_link = review_soup.find('div', {'class': 'ooJZfD _2oZ8XT col-9-12'})
```

```python
rating_list = []
review_header_list = []
detailed_review_list = []
user_list = []
likes_dislikes_list = []


for review in all_reviews_link:
    review_header = review.find_all('p', {'class': '_2xg6Ul'})
    detailed_review = review.find_all('div', {'class': 'qwjRop'})
    detailed_review=[e.get_text() for e in detailed_review]
    user = review.find_all('p', {'class': '_3LYOAd _3sxSiS'})
    user=[e.get_text() for e in user]
    likes_dislikes = review.find_all('span', {'class': '_1_BQL8'})
    likes_dislikes=[e.get_text() for e in likes_dislikes]
    rating = review.find_all('div', {'class': 'hGSR34 E_uFuv'})
    text = [e.get_text() for e in rating]
    
    
    rating_list.append(text)
    review_header=[e.get_text() for e in review_header]
    review_header_list.append(review_header)
    detailed_review_list.append(detailed_review)
    user_list.append(user)
    likes_dislikes_list.append(likes_dislikes)
```

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(list(zip(rating_list, review_header_list,detailed_review_list,user_list,likes_dislikes_list)), 
               columns =['Ratings', 'Review_Header','Detailed_Review','User','Likes/Dislikes'])
df=df.replace("[]",np.NaN)
df
```
